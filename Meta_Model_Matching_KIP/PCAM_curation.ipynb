```python
stats = {}

for key in ['val_images_0_numpy_32', 'val_images_1_numpy_32', 'test_images_0_numpy_32', 'test_images_1_numpy_32']:
    means = []
    mins = []
    maxes = []
    for image in dictionary_pcam_32_loaded[key]:
        means.append(np.mean(image))
        mins.append(np.min(image))
        maxes.append(np.max(image))
    stats[key] = {'means': means, 'mins': mins, 'maxes': maxes}

for key, value in stats.items():
    print(f"{key} Means: {np.mean(value['means'])}")
    print(f"{key} Mins: {np.mean(value['mins'])}")
    print(f"{key} Maxes: {np.mean(value['maxes'])}")
```

```python
for key, value in stats.items():
    print(f"{key} Means: {np.mean(value['means'])}")
    # plot means
    plt.hist(value['means'])
    plt.title(f"{key} Means")
    plt.show()
    print(f"{key} Mins: {np.mean(value['mins'])}")
    # plot mins
    plt.hist(value['mins'])
    plt.title(f"{key} Mins")
    plt.show()
    print(f"{key} Maxes: {np.mean(value['maxes'])}")
    # plot maxes
    plt.hist(value['maxes'])
    plt.title(f"{key} Maxes")
    plt.show()
```

```python
import matplotlib.pyplot as plt

for key in ['val_images_0_32', 'val_images_1_32', 'test_images_0_32', 'test_images_1_32']:
    images = dictionary_pcam_32_loaded[key]
    means = stats[key.replace('_32', '_numpy_32')]['means']
    sorted_indices = np.argsort(means)
    lowest_indices = sorted_indices[:5]
    highest_indices = sorted_indices[-5:]
    
    print(f"{key} Lowest Means:")
    for i in lowest_indices:
        plt.imshow(images[i])
        plt.show()
    
    print(f"{key} Highest Means:")
    for i in highest_indices:
        plt.imshow(images[i])
        plt.show()
```

```python
new_dict = {}

for key in ['val_images_0_32', 'val_images_1_32', 'test_images_0_32', 'test_images_1_32']:
    images = dictionary_pcam_32_loaded[key]
    numpy_key = key.replace('_32', '_numpy_32')
    numpy_images = dictionary_pcam_32_loaded[numpy_key]
    label_key = key.replace('images', 'labels').replace('_32', '_numpy')
    labels = dictionary_pcam_32_loaded[label_key]
    
    means = stats[numpy_key]['means']
    sorted_indices = np.argsort(means)
    n_to_exclude = int(np.ceil((len(means) - 10000) / 2))
    indices_to_keep = sorted_indices[n_to_exclude:-n_to_exclude][:10000]
    
    new_dict[key] = [images[i] for i in indices_to_keep]
    new_dict[numpy_key] = [numpy_images[i] for i in indices_to_keep]
    new_dict[label_key] = [labels[i] for i in indices_to_keep]
```

```python

for key, value in stats.items():
    print(f"{key} Means: {np.mean(value['means'])}")
    print(f"{key} Mins: {np.mean(value['mins'])}")
    print(f"{key} Maxes: {np.mean(value['maxes'])}")

stats10000 = {}

for key in ['val_images_0_numpy_32', 'val_images_1_numpy_32', 'test_images_0_numpy_32', 'test_images_1_numpy_32']:
    means = []
    mins = []
    maxes = []
    for image in new_dict[key]:
        means.append(np.mean(image))
        mins.append(np.min(image))
        maxes.append(np.max(image))
    stats10000[key] = {'means': means, 'mins': mins, 'maxes': maxes}

for key, value in stats10000.items():
    print(f"{key} Means: {np.mean(value['means'])}")
    print(f"{key} Mins: {np.mean(value['mins'])}")
    print(f"{key} Maxes: {np.mean(value['maxes'])}")
```

```python
for key, value in stats10000.items():
    print(f"{key} Means: {np.mean(value['means'])}")
    # plot means
    plt.hist(value['means'])
    plt.title(f"{key} Means")
    plt.show()
    print(f"{key} Mins: {np.mean(value['mins'])}")
    # plot mins
    plt.hist(value['mins'])
    plt.title(f"{key} Mins")
    plt.show()
    print(f"{key} Maxes: {np.mean(value['maxes'])}")
    # plot maxes
    plt.hist(value['maxes'])
    plt.title(f"{key} Maxes")
    plt.show()
```

```python
for key in ['val_images_0_32', 'val_images_1_32', 'test_images_0_32', 'test_images_1_32']:
    images = new_dict[key]
    means = stats10000[key.replace('_32', '_numpy_32')]['means']
    sorted_indices = np.argsort(means)
    lowest_indices = sorted_indices[:5]
    highest_indices = sorted_indices[-5:]
    
    print(f"{key} Lowest Means:")
    for i in lowest_indices:
        plt.imshow(images[i])
        plt.show()
    
    print(f"{key} Highest Means:")
    for i in highest_indices:
        plt.imshow(images[i])
        plt.show()
```

```python
final_dict = {}

for key in new_dict.keys():
    if key.endswith('_0_32') or key.endswith('_0_numpy_32') or key.endswith('_0_numpy'):
        new_key = key.replace('_0_', '_')
        final_dict[new_key] = np.concatenate((new_dict[key], new_dict[key.replace('_0_', '_1_')]))
print(final_dict.keys())

import matplotlib.pyplot as plt

for key, value in final_dict.items():
    first_value = value[0]
    if key in ['val_images_32', 'test_images_32']:
        plt.imshow(first_value)
        plt.show()
    else:
        print(f"{key}: {first_value}")
```
